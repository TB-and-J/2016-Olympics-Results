```python
import pandas as pd
```

```python
df = pd.read_csv("athlete_events.csv")
df.head()
```

```python
relevant = df.loc[df["Year"] == 2016]
relevant.head()
```

```python
more_relevant = relevant[["ID", "Name", "Sex", "Age", "Height", "Weight", "Team", "Sport", "Event", "Medal"]]
more_relevant.head()
```

```python
more_relevant.groupby("Medal").count()
```

```python
more_relevant['Medal'] = df["Medal"].fillna(value="None")
more_relevant.head()
```

```python
golds = more_relevant.loc[more_relevant["Medal"] == "Gold"]
golds = golds[["ID", "Medal"]]
golds.rename(columns={'Medal':'Gold Medals'}, inplace=True)
golds = golds.groupby("ID").count()
golds.head()
```

```python
silvers = more_relevant.loc[more_relevant["Medal"] == "Silver"]
silvers = silvers[["ID", "Medal"]]
silvers.rename(columns={'Medal':'Silver Medals'}, inplace=True)
silvers = silvers.groupby("ID").count()
silvers.head()
```

```python
bronzes = more_relevant.loc[more_relevant["Medal"] == "Bronze"]
bronzes = bronzes[["ID", "Medal"]]
bronzes.rename(columns={'Medal':'Bronze Medals'}, inplace=True)
bronzes = bronzes.groupby("ID").count()
bronzes.head()
```

```python
frames = [golds, silvers, bronzes]
medal_df = pd.concat(frames, sort=True)
medal_df.fillna(value = "0", inplace=True)
medal_df.head()
```
