```python
#dependencies 
import pandas as pd
from sqlalchemy import create_engine
from config import un, pw, uri, port
```

```python
#importing 120 olympics athletes csv file 
df = pd.read_csv("athlete_events.csv")
df.head()
```

```python
#only select 2016 Rio Olympics data
relevant = df.loc[df["Year"] == 2016]
relevant.head()
```

```python
more_relevant = relevant[["ID", "Name", "Sex", "Age", "Height", "Weight", "Team", "Sport", "Event", "Medal"]]
more_relevant.head()
```

```python
more_relevant = more_relevant.rename(columns= {"ID":"id", "Name":"name", "Sex":"sex", "Age":"age", 
                                               "Height":"height", "Weight":"weight", "Team":"country", 
                                               "Sport":"sport", "Event":"event","Medal":"medal"})
more_relevant.head()
```

```python
more_relevant.groupby("medal").count()
```

```python
more_relevant['medal'] = df["Medal"].fillna(value="None")
more_relevant.head()
```

```python
#building dataframe which contain gold, silver, bronze and total medal count

golds = more_relevant.loc[more_relevant["medal"] == "Gold"]
golds = golds[["id", "medal"]]
golds.rename(columns={'medal':'gold medals'}, inplace=True)
golds = golds.groupby("id").count()
golds.head()
```

```python
silvers = more_relevant.loc[more_relevant["medal"] == "Silver"]
silvers = silvers[["id", "medal"]]
silvers.rename(columns={'medal':'silver medals'}, inplace=True)
silvers = silvers.groupby("id").count()
silvers.head()
```

```python
bronzes = more_relevant.loc[more_relevant["medal"] == "Bronze"]
bronzes = bronzes[["id", "medal"]]
bronzes.rename(columns={'medal':'bronze medals'}, inplace=True)
bronzes = bronzes.groupby("id").count()
bronzes.head()
```

```python
frames = [golds, silvers, bronzes]
medal_df = pd.concat(frames, sort=True)
medal_df.fillna(value = "0", inplace=True)
medal_df.head()
```

```python
more_relevant.columns
```

```python
#importing country csv file 
country_gdp = pd.read_csv('countries.csv')
```

```python
country_gdp.head()
```

```python
country_gdp.fillna(value=0, inplace=True)
country_gdp.head()
```

```python
#remove all country which contain (*) at the end
country = country_gdp['country'].str.split(pat='*', expand=True)
```

```python
country.columns.values
```

```python
country_name = country[0]
country_name.head()
```

```python
country_gdp.head()
```

```python
country_gdp['country_name'] = country_name
country_gdp = country_gdp[['country_name', 'population', 'gdp_per_capita']]
country_gdp.head()
```

```python
country_gdp = country_gdp.rename(columns = {'country_name':'country', })
```

```python
print(country_gdp.columns)
print(more_relevant.columns)
```

```python
data_df = pd.merge(more_relevant, country_gdp, on = 'country', how = 'outer')
data_df.head()
```

```python
medal_df = medal_df.reset_index()
medal_df.head()
```

```python
final_df = pd.merge(data_df, medal_df, on='id', how='outer')
final_df.head()
```

```python
final_df = final_df.fillna({'bronze medals':0, 'gold medals':0, 'silver medals':0})
```

```python
final_df = final_df.drop(['medal'], axis =1)
```

```python
final_df.dtypes
```

```python
final_df['gold medals'] = final_df['gold medals'].astype(int)
final_df['silver medals'] = final_df['silver medals'].astype(int)
final_df['bronze medals'] = final_df['bronze medals'].astype(int)
```

```python
final_df['total medals'] = final_df['gold medals'] + final_df['silver medals'] + final_df['bronze medals']
```

```python
final_df.head()
```

```python
final_df.columns
```

```python
#building 5 new dataframes which match table names in the database
```

```python
country_df = final_df[['id','country','population','gdp_per_capita']].copy()
country_df.head()
```

```python
country_df["gdp_per_capita"] = country_df["gdp_per_capita"].convert_objects(convert_numeric=True)
```

```python
country_df["gdp_per_capita"] = round(country_df["gdp_per_capita"],2)
```

```python
sport_df = final_df[['id','sport']].copy()
sport_df.head()
```

```python
athlete_df = final_df[['id', 'name', 'sex', 'age', 'height', 'weight']].copy()
athlete_df.head()                      
```

```python
medal_df = final_df[['total medals', 'gold medals', 'silver medals', 'bronze medals']].copy()
medal_df.head()
```

```python
medal_df = medal_df.rename(columns={'total medals':'medal_count','gold medals':'gold_count',
                                    'silver medals':'silver_count','bronze medals':'bronze_count'})
medal_df.head()
```

```python
event_df = final_df[['id','sport','event']].copy()
event_df.head()
```

```python
#connect to local database 
rds_connection_string = f"{un}:{pw}@{uri}:{port}/rio_olympics_db"
engine = create_engine(f'mysql://{rds_connection_string}')
```

```python
engine.table_names()
```

```python
#use pandas to load csv converted dataframe into database 
```

```python
country_df.to_sql(name='country', con=engine, if_exists='append', index=False)
```

```python
sport_df.to_sql(name='sport', con=engine, if_exists='append', index=False)
```

```python
athlete_df.to_sql(name='athlete', con=engine, if_exists='append', index=False)
```

```python
medal_df.to_sql(name='medal', con=engine, if_exists='append', index=False)
```

```python
event_df.to_sql(name='event', con=engine, if_exists='append', index=False)
```
